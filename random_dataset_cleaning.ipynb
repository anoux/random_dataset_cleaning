```python
#supose two columns (Ref. and Alt Time Peak are useless. Then I delete it by

df = df.drop(columns = ['Ref.', 'All Time Peak'])
df
```

```python
# cleaning of columns "Peak". The last two were "cleaned" by replacing
# the actual value for the new one without the strange characters

df["Peak"] = df["Peak"].str.rstrip("[4]")

df["Peak"] = df["Peak"].str.strip("[7")

df["Peak"] = df["Peak"].str.strip("[c")

df["Peak"] = df["Peak"].str.replace('2[10','2')

df["Peak"] = df["Peak"].str.replace('1[20','1')
```

```python
# cleaning of columns "Actual gross"

df["Actual gross "] = df["Actual gross "].str.rstrip("[e]")

df["Actual gross "] = df["Actual gross "].str.rstrip("[b]")

#looks like there is a problem with the header
```
